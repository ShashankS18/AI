```
import matplotlib.pyplot as plt
```

```
x=[2,9,3,7]
y=[7,15,5,9]
plt.xlabel("Distance")
plt.ylabel("Time")
plt.plot(x,y,label="Railway Track")  # this will drwa a line
plt.grid(color="pink") # draw a of pink color
plt.scatter(x,y,label="Railway station",marker="^",color="red") #scatter plot
plt.bar(x,y,label="Altitude")
plt.legend() # display labelin the plot
plt.show()
```
